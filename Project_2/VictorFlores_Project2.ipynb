### Asynchronus Application

```python
#asynchronous modules

import asyncio
import time
import nest_asyncio
import sys
import math


sys.set_int_max_str_digits(100000000) # increases the limit of digits in integer string

nest_asyncio.apply() # allows async main to run

async def is_prime(number):
    """Is it prime number?"""
    if number<=1:
        return False
    for i in range(2,int(number**0.5)+1):
        if number % i == 0:
            return False
        
    return True

async def HI_prime():
    """Computes the highest prime number in a given amount of time 
        (3 min in this application)"""
    init_time = time.time()
    asynch_num = 0

    while time.time()-init_time<1:
        if await is_prime(asynch_num):
            HI_prime_num = asynch_num
            
        asynch_num += 1
    
    return HI_prime_num

async def fibonacci_HI_prime (hi_prime):
    """Calculates the fibonacci of the highest prime number"""
    if hi_prime <= 1:
        return hi_prime
    current, next = 0 , 1
    for _ in range(2,hi_prime + 1):
        current, next = next, current + next
        
    await asyncio.sleep(0)
    return next


async def factorial_of_HI_prime (hi_prime_num):
    """calculates the factorial of the highest prime number"""
    await asyncio.sleep(0)

    return math.factorial(hi_prime_num)


async def main():

    prime_task = asyncio.create_task(HI_prime()) #async task list 1
    prime_HI = await prime_task # outputs function output
    fib_task = asyncio.create_task(fibonacci_HI_prime(prime_HI)) # async task list 2
    fact_task = asyncio.create_task(factorial_of_HI_prime(prime_HI)) # async task list 3
    
    if prime_HI > 0:
        fib_number = await fib_task
        factorial_calc = await fact_task

    print(f"The Highest Prime Number: {prime_HI}\n\nFibonacci Number:{fib_number}\n\nFactorial:{factorial_calc}")


         

    
    
asyncio.run(main())

```

### Threading

```python
import threading
import queue
import math
import time
```

```python
def is_prime_thread(thr_number):
    """Is it prime number?"""
    if thr_number<=1:
        return False
    for i in range(2,int(thr_number**0.5)+1):
        if thr_number % i == 0:
            return False
        
    return True

def HI_prime_thread(PrimeQueue):
    """Computes the highest prime number in a given amount of time 
        (3 min in this application)"""
    init_time = time.time()
    thread_num = 0

    while time.time()-init_time< 1:
        if is_prime(thread_num):
            thread_HI_prime_num = thread_num
            
        thread_num += 1
    
    PrimeQueue.put(thread_HI_prime_num)

def fibonacci_HI_prime_thr (thr_hi_prime,FibQueue):
    """Calculates the fibonacci of the highest prime number"""
    if thr_hi_prime <= 1:
        FibQueue.put(thr_hi_prime)
    current_thr, next_thr = 0 , 1
    for _ in range(2,thr_hi_prime + 1):
        current_thr, next_thr = next_thr, current_thr + next_thr
        
    
    FibQueue.put(next_thr)


def factorial_of_HI_prime_thr (thr_hi_prime_num,FactQueue):
    """calculates the factorial of the highest prime number"""
    

    FactQueue.put(math.factorial(thr_hi_prime_num))
```

```python
if __name__ == "__main__":

    HI_prime_queue = queue.Queue()
    fibonacci_queue = queue.Queue()
    factorial_queue = queue.Queue()

    # active_threads = []

    thr1 = threading.Thread(target=HI_prime_thread,args=(HI_prime_queue))
    # active_threads.append(thr1)
    thr1.start()
    thr1.join()
    
    thr2 = threading.Thread(target=fibonacci_HI_prime_thr,args=(HI_prime_queue,fibonacci_queue))
    # active_threads.append(thr2)
    thr3 = threading.Thread(target=factorial_of_HI_prime_thr,args=(HI_prime_queue,factorial_queue))
    # active_threads.append(thr3)

    thr2.start()
    thr3.start()

    thr2.join()
    thr3.join()

    thr_HIprime_number = HI_prime_queue.get()
    thr_fibonacci_number = fibonacci_queue.get()
    thr_factorial_number = factorial_queue.get()

    # for thread in active_threads:
    #     thread.start()

    # for thread in active_threads:
    #     thread.join()
    
    print(f"The Highest Prime Number: {thr_HIprime_number}\n\nFibonacci Number:{thr_fibonacci_number}\n\nFactorial:{thr_factorial_number}")
    
    
```
